## 2つのデータセットを比較する

*   `aozorahack/aozorabunko_text` に存在しないファイルが大量に見つかる
    *   241 個もある

```python
from pathlib import Path

for path in df['path']:
  path = Path(path)
  path2 = path.parent / path.stem / (path.stem + '.txt')
  if not path2.exists():
    print(path)
```

```python
from pathlib import Path

count = 0
for path in df['path']:
  path = Path(path)
  path2 = path.parent / path.stem / (path.stem + '.txt')
  if not path2.exists():
    count += 1
print(count)
```

*   逆に， `aozorahack/aozorabunko_text` には存在するのに `aozorabunko/aozorabunko`（厳密には，そこから抽出したリスト）に存在しないものは存在するか？

```python
!find cards -type f | sort > filelist2.txt
```

```python
df2 = pd.read_csv('filelist2.txt', header=None, names=['path'])

df2
```

```python
list1 = set(df['path'])
list2 = set(df2['path'].apply(lambda path: str(Path(path).parent) + '.zip'))
len(list2 - list1)
```

```python
list2 - list1
```

*   これも大量に存在する。
*   しかし，同じ作品の異なる版で，除外済みのものかもしれない。
*   作品Noごとに集計してみる。

```python
works1 = set(Path(path).stem.split('_')[0] for path in list1)
works2 = set(Path(path).stem.split('_')[0] for path in list2)
len(works2 - works1)
```

*  2 件残った。
